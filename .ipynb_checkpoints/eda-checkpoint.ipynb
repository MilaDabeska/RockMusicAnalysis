```python
import numpy as np
import pandas as pd
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import squarify as sq
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
```

```python
dataset=pd.read_csv("history-of-rock-spotify.csv")
dataset.drop(['index', 'danceability.1'], axis=1, inplace = True)
dataset
```

## Визуелизации

```python
plt.figure(figsize = (12,8))
artist = dataset['artist'].value_counts()[:11].drop(['Various Artists'])
sns.barplot(x=artist.index, y=artist, palette="flare")
plt.xticks(rotation=70)
plt.xlabel('Artist name')
plt.ylabel('Number of songs')
plt.title('Top 10 rock artists')
plt.tight_layout()
```

```python
plt.figure(figsize = (16,8))
songs_grouped = dataset.groupby('name')['popularity'].sum().sort_values(ascending=False).head(10)
chart = sns.barplot(x=songs_grouped.index, y=songs_grouped)
plt.xticks(rotation=70)
plt.title('The most popular rock songs')
plt.tight_layout() 
```

```python
plt.figure(figsize=(14,8))
labels = ["%s\n%s" % (label) for label in zip(dataset.artist, dataset.name)]
sq.plot(sizes=dataset.name.value_counts().head(10), label=labels, alpha=.8, color=sns.color_palette("Spectral"))
plt.title('Top 10 Rock Songs and Artists')
plt.axis('off')
plt.show()
```

Дистрибуции за сите нумерички променливи.

```python
features = dataset[['release_date', 'length', 'popularity', 'danceability', 
                    'acousticness', 'energy', 'instrumentalness', 'liveness', 'loudness', 
                    'speechiness', 'tempo', 'time_signature', 'valence']]
```

```python
for i in features.columns:
    plt.hist(features[i], color = "lightblue", ec="red")
    plt.title(i)
    plt.show()
```

#### Plot data and a linear regression model fit

```python
fig,axs = plt.subplots(2,2)
fig.set_figheight(15)
fig.set_figwidth(15)
sns.regplot(x='danceability', y='popularity', ax=axs[0,1], data=dataset.head(100), color='b')
sns.regplot(x='speechiness', y='popularity', ax=axs[0,0], data=dataset.head(100), color='m')
sns.regplot(x='energy', y='popularity', ax=axs[1,0], data=dataset.head(100), color='g')
sns.regplot(x='valence', y='popularity', ax=axs[1,1], data=dataset.head(100), color='c')
plt.show()
```

Зависност помеѓу energy и popularity

```python
fig=plt.subplots(figsize=(10,10))
plt.title('Dependence between energy and popularity')
sns.regplot(x='energy', y='popularity', ci=None, data=dataset.head(100), color='olive')
sns.kdeplot(dataset.energy, dataset.popularity, color='tomato')
```

## Пирсонов коефициент на корелација
Тест статистика која ја мери статистичката врска или поврзаноста помеѓу две континуирани променливи. Познат е како најдобар метод за мерење на поврзаноста помеѓу променливите од интерес бидејќи се заснова на методот на коваријанса. Тој дава информации за големината на асоцијацијата, или корелацијата, како и насоката на врската.

```python
fig = plt.figure(figsize=(20,20))
fig.text(0.125,0.93,'Pearson\'s Correlation Coefficient', fontfamily='serif',fontsize=15, fontweight='bold')
sns.heatmap(dataset[dataset.columns.drop(['release_date', 'key'])].corr(), annot=True)
plt.show()
```

## Спирманов коефициент на корелација
Непараметриска мерка за корелација на ранг (статистичка зависност помеѓу рангирањето на две променливи). Проценува колку добро може да се опише врската помеѓу две променливи со помош на монотона функција.

```python
fig = plt.figure(figsize=(20,20))
fig.text(0.125,0.93,'Spearman\'s Coefficient', fontfamily='serif',fontsize=15, fontweight='bold')
sns.heatmap(dataset[dataset.columns.drop(['release_date', 'key'])].corr(method="spearman"), annot=True, cmap='Spectral_r')
plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
background_color = "#fbfbfb"
fig.patch.set_facecolor(background_color)

dataset.plot.area(y=['length', 'popularity', 'danceability', 'acousticness', 'energy', 'instrumentalness', 'liveness', 'loudness', 'speechiness', 
                 'tempo', 'time_signature', 'valence'], ax=plt.gca())
plt.show()
```
