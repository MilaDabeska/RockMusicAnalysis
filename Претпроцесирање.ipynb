```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
dataset=pd.read_csv("data/history-of-rock-spotify.csv")
dataset
```

&nbsp;

Преглед и типови на податоци

```python
rows, columns = dataset.shape
print(f'Редици: {rows}\nКолони: {columns}')
```

```python
dataset.info()
```

&nbsp;

Отстранување на дупликати

```python
dataset.drop(['index', 'danceability.1'], axis=1, inplace = True)
dataset.head()
```

&nbsp;

```python
dataset.rename(columns = {'length' : 'duration'}, inplace = True)
```

```python
dataset['duration'] = dataset['duration'].round(decimals = 2)
```

```python
dataset.head(3)
```

&nbsp;

Додавање на класна лабела 'hit' за да помогне подобро да ги класифицираме нашите податоци
- 1 = „е хит“
- 0 = „не е хит“

```python
dataset['hit'] = ''
```

Колку е поголема вредноста толку е попопуларна песната

```python
dataset['hit'] = pd.qcut(dataset['popularity'], q=2, labels=[0, 1]) 
dataset[['hit']].head()
```

```python
dataset.head()
```

```python
target_v = dataset['hit'].value_counts(normalize=True).round(3)
target_v
```

```python
sns.barplot(x=target_v.index, y=target_v, palette="BrBG")
```

```python
dataset.to_csv('data/rock_music.csv', index=False)
```
